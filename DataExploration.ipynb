```python
from google.colab import files
import matplotlib.pyplot as plt

# Group by 'sex' and sum 'thergrp_count' within each group for all four years
thergrp_count_gender_17 = data_17.groupby('sex')['thergrp_count'].sum()
thergrp_count_gender_18 = data_18.groupby('sex')['thergrp_count'].sum()
thergrp_count_gender_19 = data_19.groupby('sex')['thergrp_count'].sum()
thergrp_count_gender_20 = data_20.groupby('sex')['thergrp_count'].sum()

# Merge the data for 2017 and 2019 into one group named "Year1"
thergrp_count_gender_year1 = thergrp_count_gender_17 + thergrp_count_gender_19

# Merge the data for 2018 and 2020 into one group named "Year2"
thergrp_count_gender_year2 = thergrp_count_gender_18 + thergrp_count_gender_20

# Create subplots for thergrp_count by gender
fig, ax = plt.subplots(figsize=(8, 6))  # Adjust the figsize as needed

# Colors for each year
colors = ['#1f77b4', '#ff7f0e']  # Pleasant colors for Year1 and Year2

# Plot the data for Year1 and Year2 side by side
bar_width = 0.25
index = range(len(thergrp_count_gender_year1))

ax.bar([x - bar_width / 2 for x in index], thergrp_count_gender_year1, width=bar_width, label='Year1', color=colors[0])
ax.bar([x + bar_width / 2 for x in index], thergrp_count_gender_year2, width=bar_width, label='Year2', color=colors[1])

# Set the x-tick labels
ax.set_xticks(index)
ax.set_xticklabels(['Male', 'Female'])

ax.set_xlabel('Gender')
ax.set_ylabel('Therapeutic Group Count')
ax.set_title('Therapeutic Group Count by Gender for Year1 and Year2')

# Annotate each bar with the counts for Year1
for i, count in enumerate(thergrp_count_gender_year1):
    ax.text(i - bar_width / 2, count + 1, str(count), ha='center', va='bottom', fontsize=10)

# Annotate each bar with the counts for Year2
for i, count in enumerate(thergrp_count_gender_year2):
    ax.text(i + bar_width / 2, count + 1, str(count), ha='center', va='bottom', fontsize=10)

# Add legend
ax.legend()

plt.tight_layout()

# Save the plot as an image file (PNG) with high DPI
plt.savefig('thergrp_count_gender.png', dpi=300)  # Adjust the filename as needed

# Download the saved image file from Google Colab to local device
files.download('thergrp_count_gender.png')  # This will prompt you to download the file
```

```python
from google.colab import files
import matplotlib.pyplot as plt

# Group by 'agegrp' and sum 'thergrp_count' within each group for all four years
thergrp_count_2017 = data_17.groupby('agegrp')['thergrp_count'].sum()
thergrp_count_2018 = data_18.groupby('agegrp')['thergrp_count'].sum()
thergrp_count_2019 = data_19.groupby('agegrp')['thergrp_count'].sum()
thergrp_count_2020 = data_20.groupby('agegrp')['thergrp_count'].sum()

# Group by 'sex' and sum 'thergrp_count' within each group for all four years
thergrp_count_gender_2017 = data_17.groupby('sex')['thergrp_count'].sum()
thergrp_count_gender_2018 = data_18.groupby('sex')['thergrp_count'].sum()
thergrp_count_gender_2019 = data_19.groupby('sex')['thergrp_count'].sum()
thergrp_count_gender_2020 = data_20.groupby('sex')['thergrp_count'].sum()

# Merge the data for 2017 and 2019 into one group named "Year1"
thergrp_count_year1 = thergrp_count_2017 + thergrp_count_2019
thergrp_count_gender_year1 = thergrp_count_gender_2017 + thergrp_count_gender_2019

# Merge the data for 2018 and 2020 into one group named "Year2"
thergrp_count_year2 = thergrp_count_2018 + thergrp_count_2020
thergrp_count_gender_year2 = thergrp_count_gender_2018 + thergrp_count_gender_2020

# Create separate plots for therapeutic group count by age group and by gender
fig, axes_age = plt.subplots(figsize=(8, 5))
fig, axes_gender = plt.subplots(figsize=(8, 5))

# Define colors for each year
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Pleasant colors for each year

# Increase the font size for labels and annotations
plt.rcParams.update({'font.size': 12})

# Plot therapeutic group count by age group
axes_age.bar(thergrp_count_year1.index.astype(int) - 0.2, thergrp_count_year1, width=0.4, label='Year1', color=colors[0])
axes_age.bar(thergrp_count_year2.index.astype(int) + 0.2, thergrp_count_year2, width=0.4, label='Year2', color=colors[1])
axes_age.set_xlabel('Age Group')
axes_age.set_ylabel('Therapeutic Group Count')
axes_age.set_title('Therapeutic Group Count by Age Group for All Years')

# Annotate each bar with the counts for Year1
for i, count in enumerate(thergrp_count_year1):
    axes_age.text(i - 0.2, count + 2, str(count), ha='center', va='bottom', fontsize=10)

# Annotate each bar with the counts for Year2
for i, count in enumerate(thergrp_count_year2):
    axes_age.text(i + 0.2, count + 2, str(count), ha='center', va='bottom', fontsize=10)

# Plot therapeutic group count by gender
bar_width = 0.35
index = range(len(thergrp_count_gender_year1))

axes_gender.bar([x - bar_width / 2 for x in index], thergrp_count_gender_year1, width=bar_width, label='Year1', color=colors[0])
axes_gender.bar([x + bar_width / 2 for x in index], thergrp_count_gender_year2, width=bar_width, label='Year2', color=colors[1])
axes_gender.set_xticks(index)
axes_gender.set_xticklabels(['Male', 'Female'])
axes_gender.set_xlabel('Sex')
axes_gender.set_ylabel('Therapeutic Group Count')
axes_gender.set_title('Therapeutic Group Count by Sex for All Years')
axes_gender.legend()

# Annotate each bar with the counts for Year1
for i, count in enumerate(thergrp_count_gender_year1):
    axes_gender.text(i - bar_width / 2, count + 2, str(count), ha='center', va='bottom', fontsize=10)

# Annotate each bar with the counts for Year2
for i, count in enumerate(thergrp_count_gender_year2):
    axes_gender.text(i + bar_width / 2, count + 2, str(count), ha='center', va='bottom', fontsize=10)

# Save the plots as image files (PNG) with high DPI
plt.savefig('therapeutic_group_count_age_hd_labels.png', dpi=300, bbox_inches='tight')
plt.savefig('therapeutic_group_count_gender_hd_labels.png', dpi=300, bbox_inches='tight')

# Download the saved image files from Google Colab to local device
files.download('therapeutic_group_count_age_hd_labels.png')  # This will prompt you to download the file
files.download('therapeutic_group_count_gender_hd_labels.png')  # This will prompt you to download the file

plt.show()
```

```python
from google.colab import files
import matplotlib.pyplot as plt

# Group by 'agegrp' and sum 'thergrp_count' within each group for all four years
thergrp_count_2017 = data_17.groupby('agegrp')['thergrp_count'].sum()
thergrp_count_2018 = data_18.groupby('agegrp')['thergrp_count'].sum()
thergrp_count_2019 = data_19.groupby('agegrp')['thergrp_count'].sum()
thergrp_count_2020 = data_20.groupby('agegrp')['thergrp_count'].sum()

# Merge the data for 2017 and 2019 into one group named "Year1"
thergrp_count_year1 = thergrp_count_2017 + thergrp_count_2019

# Merge the data for 2018 and 2020 into one group named "Year2"
thergrp_count_year2 = thergrp_count_2018 + thergrp_count_2020

# Create a plot for therapeutic group count by age group
fig, axes_age = plt.subplots(figsize=(8, 5))

# Define colors for each year
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Pleasant colors for each year

# Increase the font size for labels and annotations
plt.rcParams.update({'font.size': 12})

# Plot therapeutic group count by age group
bars1 = axes_age.bar(thergrp_count_year1.index.astype(int) - 0.2, thergrp_count_year1, width=0.4, label='Year1', color=colors[0])
bars2 = axes_age.bar(thergrp_count_year2.index.astype(int) + 0.2, thergrp_count_year2, width=0.4, label='Year2', color=colors[1])

# Annotate each bar with the counts for Year1 and Year2
for bar, count in zip(bars1, thergrp_count_year1):
    plt.annotate('{}k'.format(int(count/1000)), xy=(bar.get_x() + bar.get_width()/2, bar.get_height()), xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')
for bar, count in zip(bars2, thergrp_count_year2):
    plt.annotate('{}k'.format(int(count/1000)), xy=(bar.get_x() + bar.get_width()/2, bar.get_height()), xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

axes_age.set_xlabel('Age Group')
axes_age.set_ylabel('Therapeutic Group Count')
axes_age.set_title('Therapeutic Group Count by Age Group for All Years')

# Save the plot as an image file (PNG) with high DPI
plt.savefig('therapeutic_group_count_age_hd_labels.png', dpi=300, bbox_inches='tight')

# Download the saved image file from Google Colab to local device
files.download('therapeutic_group_count_age_hd_labels.png')  # This will prompt you to download the file

plt.show()
```
