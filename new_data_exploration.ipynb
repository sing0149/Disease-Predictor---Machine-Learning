```python
# FULL-DATA
# specify which column to read, so that the unneccessay column don't take up memory
columns = ['ENROLID','YEAR', 'AGE', 'DX1', 'DX2',
           'DX3', 'DX4', 'DX5', 'DX6', 'DX7',
           'DX8', 'DX9', 'DX10', 'DX11', 'DX12',
           'DX13', 'DX14', 'DX15','SEX']


ccaei192_a = pd.read_csv('ccaei192_a.csv',
                 skiprows = range(1, 3855),  # skiprows = 3855, before that all Enrolid is NaN
                 usecols = columns)

ccaei192_a.head()
```

### Load in 2020 Inpatient Admission Dataset

```python
ccaei200_a = pd.read_csv('ccaei200_a.csv',
                         skiprows = range(1, 3458),  # skip the missing Enrolid
                         usecols = columns)

ccaei200_a.head()
```

### Merge Two Inpatient Admission Dataset Together

```python
ccaei19_20 = pd.merge(ccaei192_a, ccaei200_a,
                      on='ENROLID',  # merge on EnrolID to get the Ids that occurs in both years
                      how='inner')   # inner: use intersection of keys from both frames, similar to a SQL inner join; preserve the order of the left keys.
```

```python
ccaei19_20.head()
```

```python
ccaei19_20.shape
```

```python
# drop duplicate rows (rows that have exactly the same information due to an earlier inner join)
ccaei19_20 = ccaei19_20.drop_duplicates()
```

```python
ccaei19_20.shape
```
